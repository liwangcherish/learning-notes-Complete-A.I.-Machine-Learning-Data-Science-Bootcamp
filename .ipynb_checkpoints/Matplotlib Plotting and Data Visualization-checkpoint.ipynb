## Matplotlib Example Workflow 

```python
%matplotlib inline
import matplotlib.pyplot as plt

# 1. prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. setup plot
fig, ax = plt.subplots(figsize = (10, 10)) #width, height

# 3 plot data
ax.plot(x, y)

# 4. customizr plot
ax.set(title = "Simple Plot",
       xlabel = "x- axis",
       ylabel = "y-axis")

# 5. Save & show
fig.savefig("sample-plot.png");
```

## Making figures with Numpy

We want:
* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplots

```python
import numpy as np
```

```python
# Create some data
x = np.linspace(0, 10, 100)
x[:10]
```

```python
# plot the data
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# Use same data to make a scatter
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Make a plot from dictionary
nut_butter_prices = {"Almond butter": 10,
                     "Peanut butter": 8,
                     "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values()); # x, y
ax.set(title = "Dan's Nut Butter Store",
       ylabel = "Price($)")
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

```python
# Make some data for histogram and plot it
x = np.random.randn(10000)
fig, ax = plt.subplots()
ax.hist(x);
```

# Two options for subplots

```python
# Subplots Option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2,
                                             ncols = 2,
                                             figsize = (10, 5))
#plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# subplots option 2
fig, ax = plt.subplots(nrows = 2,
                       ncols = 2,
                       figsize = (10, 5))
# plot to each diffrent index
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

```python
car_sales
```

```python
# Data manipulate for price $4,000,00 -> 4000
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\.]', '')
car_sales
```

```python
# remove last two zeros 
type(car_sales["Price"][0])
car_sales["Price"] = car_sales["Price"].str[:-2]
car_sales
```

```python
car_sales["Sale Date"] = pd.date_range('1/1/2020', periods = len(car_sales))
car_sales
```

```python
car_sales["Total Sales"] = car_sales["Price"].cumsum()
car_sales
```

```python
type(car_sales["Price"][0])
```

```python
#need to change price to integer
car_sales["Total Sales"] = car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
# plot total sales 
```

```python
car_sales.plot(x = "Sale Date", y = "Total Sales");
```

```python
# assign price colume to integer
car_sales["Price"] = car_sales["Price"].astype(int)
car_sales.plot(x = "Odometer (KM)", y = "Price", kind = "scatter");
```

```python
car_sales["Price"].plot();
```

```python
# average_price datatype is series, like a column, it should be y
average_price = car_sales.groupby(["Make"])["Price"].mean()
average_price
```

```python
average_price.plot(x = "Make", y = "average_price", kind = "bar", title = 'Average Car Price by Make');
```

```python
car_sales
```

```python
#  average price by colour
average_price_by_colour = car_sales.groupby(["Colour"])["Price"].mean()
average_price_by_colour.plot(x = "Colour", y = average_price_by_colour, kind = "bar");
```
