## Matplotlib Example Workflow 

```python
%matplotlib inline
import matplotlib.pyplot as plt

# 1. prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. setup plot
fig, ax = plt.subplots(figsize = (10, 10)) #width, height

# 3 plot data
ax.plot(x, y)

# 4. customizr plot
ax.set(title = "Simple Plot",
       xlabel = "x- axis",
       ylabel = "y-axis")

# 5. Save & show
fig.savefig("sample-plot.png");
```

## Making figures with Numpy

We want:
* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplots

```python
import numpy as np
```

```python
# Create some data
x = np.linspace(0, 10, 100)
x[:10]
```

```python
# plot the data
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# Use same data to make a scatter
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Make a plot from dictionary
nut_butter_prices = {"Almond butter": 10,
                     "Peanut butter": 8,
                     "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values()); # x, y
ax.set(title = "Dan's Nut Butter Store",
       ylabel = "Price($)")
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

```python
# Make some data for histogram and plot it
x = np.random.randn(10000)
fig, ax = plt.subplots()
ax.hist(x);
```
