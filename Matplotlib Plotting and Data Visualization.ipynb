## Matplotlib Example Workflow 

```python
%matplotlib inline
import matplotlib.pyplot as plt

# 1. prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. setup plot
fig, ax = plt.subplots(figsize = (10, 10)) #width, height

# 3 plot data
ax.plot(x, y)

# 4. customizr plot
ax.set(title = "Simple Plot",
       xlabel = "x- axis",
       ylabel = "y-axis")

# 5. Save & show
fig.savefig("sample-plot.png");
```

# Two options for subplots

```python
# Subplots Option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2,
                                             ncols = 2,
                                             figsize = (10, 5))
#plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# subplots option 2
fig, ax = plt.subplots(nrows = 2,
                       ncols = 2,
                       figsize = (10, 5))
# plot to each diffrent index
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

## Which one should you use (pyplot vs marplotlib OO method?)
* pyplot , something quickly 
* more advanced, use OO method 

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
#pyplot method, pandas
%matplotlib notebook

over_50.plot(kind = 'scatter',
            x = 'age',
            y = 'chol',
            c = 'target',
            figsize = (10,6
                    ));
```

```python
# OO method mixed with pyplot
fix, ax = plt.subplots(figsize = (10, 6))
over_50.plot(kind = 'scatter',
            x = 'age',
            y = 'chol',
            c = 'target',
            ax = ax);

# add limitations of the axis
# ax.set_xlim([45, 100]);
```

```python
# OO method from scratch 
fig, ax = plt.subplots(figsize = (10, 10))

# plot the data
scatter = ax.scatter(x = over_50["age"],
                    y = over_50["chol"],
                    c = over_50["target"])

# Customize
ax.set(title = "Heart Disease and Cholesterol Levels",
       xlabel = 'Age',
       ylabel = 'Cholesterol');

# Add a legend
ax.legend(*scatter.legend_elements(), title='Target');

# Add a horizontal line
ax.axhline(over_50['chol'].mean(),
          linestyle = '--');
```

```python
over_50.head()
```

```python
# Object Oriented Plot

# subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows = 2,
                              ncols = 1,
                              figsize = (10, 10),
                              sharex=True);

# add data to ax0
scatter = ax0.scatter(x = over_50["age"],
                     y = over_50["chol"],
                     c = over_50["target"]);

#customize ax0
ax0.set(title = "Heart Disease and Cholesterol Levels",
       ylabel = "Cholesterol");

# Add a ledeng to ax0
ax0.legend(*scatter.legend_elements(), title = "Target")
# Add a meanline
ax0.axhline(y = over_50["chol"].mean(),
           linestyle = "--");

# add data to ax1
scatter = ax1.scatter(x = over_50["age"],
                     y = over_50["thalach"],
                     c = over_50["target"]);

#customize ax1
ax1.set(title = "Heart Disease and thalach",
       xlabel = "Age",
       ylabel = "Thalach");

# Add a ledeng to ax1
ax1.legend(*scatter.legend_elements(), title = "Target")
# Add a meanline
ax1.axhline(y = over_50["thalach"].mean(),
           linestyle = "--");

# Add a title into the figure
fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight = "bold");
```

# sample 

```python
# customize the y and x axis limitations

# Object Oriented Plot

# subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows = 2,
                              ncols = 1,
                              figsize = (10, 10),
                              sharex=True);

# add data to ax0
scatter = ax0.scatter(x = over_50["age"],
                     y = over_50["chol"],
                     c = over_50["target"],
                     cmap = "winter");

#customize ax0
ax0.set(title = "Heart Disease and Cholesterol Levels",
       ylabel = "Cholesterol");

# change the x axis limit 
ax0.set_xlim([50, 80])

# Add a ledeng to ax0
ax0.legend(*scatter.legend_elements(), title = "Target")
# Add a meanline
ax0.axhline(y = over_50["chol"].mean(),
           linestyle = "--");

# add data to ax1
scatter = ax1.scatter(x = over_50["age"],
                     y = over_50["thalach"],
                     c = over_50["target"],
                     cmap = "winter");

#customize ax1
ax1.set(title = "Heart Disease and thalach",
       xlabel = "Age",
       ylabel = "Thalach");
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])
# Add a ledeng to ax1
ax1.legend(*scatter.legend_elements(), title = "Target")
# Add a meanline
ax1.axhline(y = over_50["thalach"].mean(),
           linestyle = "--");

# Add a title into the figure
fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight = "bold");
```
