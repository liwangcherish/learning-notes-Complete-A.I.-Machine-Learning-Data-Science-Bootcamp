# 0. An end-to-end Scikit-Learn workflow

```python
# Standard imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 1. Get the data ready
import pandas as pd

# specify the file path
path = '/Users/babilibabyli/Documents/ds/sample_project/zero-to-mastery-ml-master/learning-notes-Complete-A.I.-Machine-Learning-Data-Science-Bootcamp/data/heart-disease.csv'

# read the csv file
heart_disease = pd.read_csv(path)
heart_disease
```

```python
# Create X (features matrix)
X = heart_disease.drop("target", axis = 1)

# Create y (labels)
y = heart_disease["target"]
```

```python
# 2. Choose the right models and hyperparameters
# classification 
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators = 100)

# We'll keep the default hyperparameters
clf.get_params()
```

```python
# Fit the model to the training data; 80% data for training, 20% for testing 
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
clf.fit(X_train, y_train);
```

```python
X_test
```

```python
# make a prediction 
y_preds = clf.predict(X_test)
y_preds
```

```python
y_test
```

```python
# 4. Evaluate the model on the training data and test data
clf.score(X_train, y_train)
```

```python
# 83% accuracy 
clf.score(X_test, y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score 

print(classification_report(y_test, y_preds))
```

```python
confusion_matrix(y_test, y_preds)
```

```python
accuracy_score(y_test, y_preds)
```

```python
# 5. Improve a model
# Try different amount of n_estimators
np.random.seed(42)
for i in range(10, 100, 10):
    print(f"Trying model with {i} estimators...")
    clf = RandomForestClassifier(n_estimators = i).fit(X_train, y_train)
    print(f"Model accuracy on test set: {clf.score(X_test, y_test) * 100: .2f}")
    print("")
```

```python
# 6. Save a model and load it 
import pickle

pickle.dump(clf, open("random_forest_model_1.pk1", "wb"))
```

```python
loaded_model = pickle.load(open("random_forest_model_1.pk1", "rb"))
loaded_model.score(X_test, y_test)
```

## 1.1 Make sure it's all numerical

```python
car_sales = pd.read_csv("data/car-sales-extended.csv")
car_sales.head()
```

```python
len(car_sales)
```

```python
car_sales.dtypes
```

```python
# Split into X/y
X = car_sales.drop("Price", axis = 1)
y = car_sales["Price"]

# Split into training and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
# Build Machine Learning Model
# Machine Learning couldn't deal with strings
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
X.head()
```

```python
# Trun the categories into numbers
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

categorical_features = ["Make", "Colour", "Doors"]
one_hot = OneHotEncoder()
transformer = ColumnTransformer([("one_hot",
                                 one_hot,
                                 categorical_features)],
                               remainder = "passthrough")
transformed_X = transformer.fit_transform(X)
transformed_X
```

```python
X.head()
```

```python
pd.DataFrame(transformed_X)
```

```python
dummies = pd.get_dummies(car_sales[["Make", "Colour", "Doors"]])
dummies
```

```python
# Let's refit the model
np.random.seed(42)
X_train, X_test, y_train, y_test = train_test_split(transformed_X,
                                                   y,
                                                   test_size = 0.2)
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
